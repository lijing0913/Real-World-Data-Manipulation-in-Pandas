**Import the necessary libraries**

```python
import pandas as pd
```

**Import the dataset from internet, and assign it to a variable**

```python
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/Students_Alcohol_Consumption/student-mat.csv'

df = pd.read_csv(url)
df.head()
```

**Slice the dataframe from 'school' until the 'guardian' column**

```python
stud_alcohol = df.loc[:, 'school': 'guardian']
stud_alcohol.head()
```

<font color=blue>**Create a lambda function that will capitalize strings**

```python
capitalizer = lambda x: x.capitalize()
```

**Capitalize both Mjob and Fjob**

```python
stud_alcohol.Mjob.apply(capitalizer)
stud_alcohol.Fjob.apply(capitalizer)
# NOTE: stud_alcohol doesn't change

# or
# stud_alcohol['Mjob'].apply(capitalizer)
# stud_alcohol['Fjob'].apply(capitalizer)
```

**Print the last elements of the data set**

```python
stud_alcohol.tail()
```

**The original dataframe is still lowercase? Why is that? Fix it and capitalize Mjob and Fjob**

```python
stud_alcohol['Mjob'] = stud_alcohol['Mjob'].apply(capitalizer)
stud_alcohol['Fjob'] = stud_alcohol['Fjob'].apply(capitalizer)
stud_alcohol.head()
```

**Create a function called majority that returns a boolean value to a new column called legal_drinker (Consider majority as older than 17 years old)**

```python
def majority(x):
    return True if x > 17 else False 
```

```python
stud_alcohol['legal_drinker'] = stud_alcohol['age'].apply(majority)
stud_alcohol.head()
```

<font color=blue>**Multiply every number of the dataset by 10**

```python
stud_alcohol.info()
```

```python
def multiply(x):
    return x * 10 if type(x) == int else x # NOTE: here == int instead of == 'int'
```
