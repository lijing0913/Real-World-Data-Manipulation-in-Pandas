# Scores

**Import the necessary libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

**Create the DataFrame that should look like the one below**

```python
raw_data = {
    'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
    'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
    'age': [42, 52, 36, 24, 73], 
    'female': [0, 1, 1, 0, 1],
    'preTestScore': [4., 24, 31, 2, 3], 
    'postTestScore': [25, 94, 57, 62, 70]
    }

df = pd.DataFrame.from_dict(raw_data, orient='columns')
# or
# df = pd.DataFrame(raw_data, columns = ['first_name', 'last_name', 'age', 'female', 'preTestScore', 'postTestScore'])
df
```

**Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age**

```python
# sns.scatterplot(
#     x=df.preTestScore, 
#     y=df.postTestScore, 
#     s=df.age)
# or
plt.scatter(x=df.preTestScore, y=df.postTestScore, s=df.age)

plt.xlabel('preTestScore')
plt.ylabel('postTestScore')
plt.title('postTestScore vs preTestScore')
plt.show()
```

**Create a Scatterplot of preTestScore and postTestScore. This time the size should be 4.5 times the postTestScore and the color determined by sex**

```python
sns.scatterplot(
    x=df.preTestScore, 
    y=df.postTestScore, 
    hue=df.female,
    s=df.age*4.5)
# or
# plt.scatter(x=df.preTestScore, y=df.postTestScore, s=df.age*4.5, c=df.female)

plt.xlabel('preTestScore')
plt.ylabel('postTestScore')
plt.title('postTestScore vs preTestScore')
plt.show()
```
