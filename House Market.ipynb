# House Market

**Import the necessary libraries**

```python
import pandas as pd
import numpy as np
```

<font color=blue>**Create 3 differents Series, each of length 100, as follows:**
1. The first a random number from 1 to 4 
2. The second a random number from 1 to 3
3. The third a random number from 10,000 to 30,000

```python
series1 = pd.Series(np.random.randint(1, 5, 100))
series2 = pd.Series(np.random.randint(1, 4, 100))
series3 = pd.Series(np.random.randint(10000, 30001, 100))

print(series1, series2, series3)
```

**Create a DataFrame by joinning the Series by column**

```python
df = pd.concat([series1, series2, series3], axis=1)
df
```

**Change the name of the columns to bedrs, bathrs, price_sqr_meter**

```python
df.columns = ['bedrs', 'bathrs', 'price_sqr_meter']

# or
# housemkt.rename(columns = {0: 'bedrs', 1: 'bathrs', 2: 'price_sqr_meter'}, inplace=True)

df
```

<font color=blue>**Create a one column DataFrame with the values of the 3 Series and assign it to 'bigcolumn'**

```python
bigcolumn = pd.concat([series1, series2, series3], axis=0)

# Convert Series to DataFrame
bigcolumn = bigcolumn.to_frame()
print(type(bigcolumn))
bigcolumn
```

**It seems it is going only until index 99. Is it true?**

```python
# no the index are kept but the length of the DataFrame is 300
len(bigcolumn)
```

```python
bigcolumn.reset_index(drop=True, inplace=True)
bigcolumn
```
