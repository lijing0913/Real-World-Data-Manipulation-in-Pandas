This dataset is from [Kaggle](https://www.kaggle.com/c/titanic)

**Import the necessary libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

**Import the dataset from the internet and assign it to a varible**

```python
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv'
titanic = pd.read_csv(url)
titanic.head()
```

**Set PassengerId as the index**

```python
titanic.set_index('PassengerId', inplace=True)
titanic.head()
```

<font color=blue><font color=blue>**Create a pie chart presenting the male/female proportion**

```python
# count the number of male and female passengers
# male_count = titanic[titanic['Sex'] == 'male'].shape[0]
# or
male_count = (titanic['Sex'] == 'male').sum()
print(male_count)

female_count = (titanic['Sex'] == 'female').sum()
print(female_count)

# create a pie chart
plt.pie(
    [male_count, female_count], 
    labels = ['Males', 'Females'],
    shadow = False,
    colors = ['blue', 'red'],
    explode = (0.15, 0), # with one slide exploded out
    startangle = 90, # with the start angle at 90 degrees
    autopct = '%1.1f%%' # with the percent listed as a fraction
    )
plt.title('Sex Proportion')
plt.tight_layout()
plt.show()
```

**Create a scatterplot with the Fare payed and the Age, differ the plot color by gender**

```python
plt.title('Fare vs Age')
sns.scatterplot(
    x=titanic.Age,
    y=titanic.Fare,
    hue=titanic.Sex
    );


# # or
# # creates the plot using
# lm = sns.lmplot(x = 'Age', y = 'Fare', data = titanic, hue = 'Sex', fit_reg=False)

# # set title
# lm.set(title = 'Fare x Age')

# # get the axes object and tweak it
# axes = lm.axes
# axes[0,0].set_ylim(-5,)
# axes[0,0].set_xlim(-5,85)
```

**How many people survived?**

```python
survived_count = titanic['Survived'].sum()
survived_percentage = survived_count / titanic.shape[0] * 100
print('There were {} people survived, which is {:.2f}%'.format(survived_count, survived_percentage))
```

**Create a histogram with the Fare payed**

```python
plt.hist(titanic.Fare, bins=np.arange(0, 600, 10));
plt.xlabel('Fare');
plt.ylabel('Frequency')
plt.title('Fare Payed Histogram');
```
