**Import the necessary libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# display the graphs in the notebook
%matplotlib inline 

# set seaborn style to while
sns.set_style('white')
```

**Import the dataset from the internet, and assign it to a variable**

```python
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv'

tips = pd.read_csv(url)
tips.head()
```

### Step 4. Delete the Unnamed 0 column

```python
del tips['Unnamed: 0']
# or
# tips.drop('Unnamed: 0', axis=1, inplace=True)
tips.head()
```

**Plot the total_bill column histogram**

```python
plt.hist(tips.total_bill)

# sns.distplot(tips.total_bill)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Total Bill Histogram')

# take out the right and upper borders
sns.despine()
```

**Create a scatter plot presenting the relationship between total_bill and tip**

```python
# sns.scatterplot(x=tips.total_bill, y=tips.tip);

# plt.scatter(tips.total_bill, tips.tip);

sns.jointplot(x=tips.total_bill, y=tips.tip);
```

<font color=blue>**Create one image with the relationship of total_bill, tip and size**

```python
sns.pairplot(tips);
```

<font color=blue>**Present the relationship between days and total_bill value**

```python
sns.stripplot(x=tips.day, y=tips.total_bill, jitter=True);
```

**Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex**

```python
sns.scatterplot(x=tips.tip, y=tips.day, hue=tips.sex);

# or
# sns.stripplot(x = "tip", y = "day", hue = "sex", data = tips, jitter = True);
```

<font color=blue>**Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)**

```python
sns.boxplot(x='day', y='total_bill', hue='time', data=tips);
```

<font color=blue>**Create two histograms of the tip value based for Dinner and Lunch. They must be side by side**

```python
# better seaborn styple
sns.set(style = 'ticks')

# create FacetGrid
fig = sns.FacetGrid(tips, col='time')
fig.map(plt.hist, 'tip')
```

<font color=blue>**Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker. They must be side by side**

```python
fig = sns.FacetGrid(tips, col='sex', hue='smoker')
fig.map(plt.scatter, 'total_bill', 'tip', alpha=0.7)
fig.add_legend();
```
