# Iris

**Dataset**: https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data

**Import the necessary libraries**

```python
import pandas as pd
import numpy as np
```

**Import the dataset from internet. Assign it to a variable**

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'

iris = pd.read_csv(url)
iris
```

**Create columns for the dataset**

```python
# 1. sepal_length (in cm)
# 2. sepal_width (in cm)
# 3. petal_length (in cm)
# 4. petal_width (in cm)
# 5. class

iris.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
iris
```

**Is there any missing value in the dataframe?**

```python
iris.isnull().sum()

# no missing values
```

**Lets set the values of the rows 10 to 29 of the column 'petal_length' to NaN**

```python
iris.loc[10:30, 'petal_length'] = np.nan
iris.head(20)
```

<font color=blue>**Good, now lets substitute the NaN values to 1.0**

```python
iris.petal_length.fillna(1, inplace=True)
iris.head(20)
```

**Now let's delete the column class**

```python
del iris['class']

# or
# iris.drop('class', axis=1, inplace=True)
iris.head()
```

**Set the first 3 rows as NaN**

```python
iris.iloc[:3, :] = np.nan
iris.head()
```

**Delete the rows that have NaN**

```python
iris.dropna(axis=0, how='any', inplace=True)
iris.head()
```

<font color=blue>**Reset the index so it begins with 0 again**

```python
iris = iris.reset_index(drop = True) ###
iris.head()
```
