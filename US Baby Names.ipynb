# US Baby Names

**Dataset**: https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv.
The data contains names from 2004 to 2014.

**Import the necessary libraries**

```python
import pandas as pd
```

**Import the dataset from the internet, and assign it to a variable**

```python
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv'

baby_names = pd.read_csv(url)
baby_names
```

**See the first 10 entries**

```python
baby_names.head(10)
```

**Delete the column 'Unnamed: 0' and 'Id'**

```python
baby_names.drop(['Unnamed: 0', 'Id'], axis=1, inplace=True)

# or
# # deletes Unnamed: 0
# del baby_names['Unnamed: 0']

# # deletes Unnamed: 0
# del baby_names['Id']

baby_names.head()
```

**Is there more male or female names in the dataset?**

```python
baby_names.Gender.value_counts()
```

**Group the dataset by name and assign to names**

```python
# Delete the Year column
del baby_names['Year']
```

```python
# Group the data
names = baby_names.groupby('Name').sum()
print(names.shape)

# Sort in descending order
names.sort_values('Count', ascending=False).head(30)
```

**How many different names exist in the dataset?**

```python
baby_names['Name'].nunique()

# or
# len(names)
```

**What is the name with most occurrences?**

```python
names.Count.idxmax()

# or
# names[names.Count == names.Count.max()]
```

<font color=blue>**How many different names have the least occurrences?**

```python
names[names.Count == names.Count.min()]
```

**What is the median name occurrence?**

```python
names[names.Count == names.Count.median()]
```

**What is the standard deviation of names?**

```python
names.Count.std()
```

**Get a summary with the mean, min, max, std and quartiles**

```python
names.describe()
```
